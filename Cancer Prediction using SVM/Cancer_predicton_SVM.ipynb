```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Breast_Cancer_Diagnostic.csv')
df
```

***Test-Train div***

```python
X=df.drop('diagnosis',axis=1)
y=df.diagnosis
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

***Manualy***

```python
svm=SVC(kernel='linear',C=30,gamma='auto') #kernel{'linear','ploy','rbf','sigmoid','precomputed or callable'}
svm.fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

***Hyperparameter Tuning gives the Best output***

**Use GridSearchCV for Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(SVC(gamma='auto'),{'C':[10,20,30],
                                   'kernel':['rbf','linear']},cv=5)

clf.fit(X,y)

clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
```

```python
df
```

```python
clf.best_params_
```

```python
svm.score(X_test,y_test)
```
