***Using Decision Tree***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('shop data.csv')
df
```

***Separate Tergated value***

```python
#Remove Tergeted Value
X=df.drop('buys',axis=1)
```

```python
y=df.buys
```

```python
y
```

```python
X
```

***Encoding***

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
X=X.apply(le.fit_transform)
```

```python
X
```

***Split Test Train data***

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
X_test
```

```python
X_train
```

```python
#Here doing math of entropy and IG
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

```python
tree.predict(X_test)
```

```python
tree.score(X_test,y_test)
```

***Using cross_val_score for better result*** 

```python
from sklearn.model_selection import cross_val_score
```

```python
result =cross_val_score(tree,X,y,cv=5)  #5 times
```

```python
result
```

```python
result.mean()
```

#### Using LOgisticRegression 

```python
from sklearn.linear_model import LogisticRegression
result =cross_val_score(LogisticRegression(),X,y,cv=5)
```
